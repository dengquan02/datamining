```python
def calc_DI(dist_matrix, labels, clusters):
    DI = 999
    k = len(clusters)
    # step 1
    max_diam = 0
    for l in range(k):
        groups = clusters[l]
        for a in groups:
            for b in groups:
                if a!=b:
#                     print(a, b)
                    max_diam = max(max_diam, dist_matrix.tolist()[a][b])
    # step 2
    for i in range(k):
        the_min = 999
        for j in range(k):
            if i != j:
                d_min = 999
                for a in clusters[i]:
                    for b in clusters[j]:
                        d_min = min(d_min, dist_matrix.tolist()[a][b])
                the_min = min(the_min, d_min / max_diam)
        DI = min(DI, the_min)
    
    return DI
```
