```python
# 给b、c用
class RoadGrid:
    def __init__(self, label):
        label = pd.DataFrame(label)
        length = 0.005
        width = 0.005

        xl, xr, yb, yt = -0.02, +0.02, -0.03, +0.03
        lenNum = int((xr-xl)/length)
        widNum = int((yt-yb)/width)
        print(lenNum,widNum)
        grid_dict = {}
        
        for m in range(0,lenNum):
            for n in range(0,widNum):
                if (m, n) not in grid_dict:
                    grid_dict[(m, n)] = np.zeros(shape=(7, 5))
                    
        for index, pos in label.iterrows():
            pos = pos.T
            lon = pos["Lon_Distance"]
            lat = pos["Lat_Distance"]

            m = int((lon-xl)/length)
            n = int((lat-yb)/width)

            for i in range(1, 8):
                if pos["SignalLevel_"+str(i)] == -999 or pos["SignalLevel_"+str(i)] == 6188:
                    continue
                j = pos["SignalLevel_"+str(i)]
#                 print(m, n)
                grid_dict[(m, n)][int(i-1)][int(j)] += 1

        self.grids = grid_dict
```

```python
weights = [
    [1],
    [0.6, 0.4],
    [0.5, 0.3, 0.2],
    [0.4, 0.3, 0.2, 0.1],
    [0.4, 0.2, 0.2, 0.1, 0.1],
    [0.3, 0.2, 0.2, 0.1, 0.1, 0.1],
    [0.25, 0.2, 0.2, 0.1, 0.1, 0.1, 0.05],
] 

# weights = [
#     [1],
#     [0.5, 0.5],
#     [1/3, 1/3, 1/3],
#     [1/4, 1/4, 1/4, 1/4],
#     [1/5, 1/5, 1/5, 1/5,1/5],
#     [1/6, 1/6, 1/6, 1/6,1/6,1/6],
#     [1/7, 1/7, 1/7, 1/7,1/7,1/7,1/7],
# ] 

weights
def dis_gird(a,b):
    D = []
    t = 0
    for i in range(7):
        if np.sum(a[i]) != 0 or np.sum(b[i]) != 0:
            t += 1
        tmp = 0
        for j in range(5):
            tmp += (a[i][j] - b[i][j])**2
        D.append(tmp**(0.5))
    dis = 0
    for k in range(len(weights[t-1])):
        dis += weights[t-1][k] * D[k]
    return dis
```

```python
def dis_group(a, b):
    D = []
    for key in a.keys():
        if key in b.keys():
            D.append(dis_gird(a[key], b[key]))
    dis = np.mean(D)
    return dis
```

```python
def calc_DI(dist_matrix, labels, clusters):
    DI = 999
    k = len(clusters)
    # step 1
    max_diam = 0
    for l in range(k):
        groups = clusters[l]
        for a in groups:
            for b in groups:
                if a!=b:
#                     print(a, b)
                    max_diam = max(max_diam, dist_matrix.tolist()[a][b])
    # step 2
    for i in range(k):
        the_min = 999
        for j in range(k):
            if i != j:
                d_min = 999
                for a in clusters[i]:
                    for b in clusters[j]:
                        d_min = min(d_min, dist_matrix.tolist()[a][b])
                the_min = min(the_min, d_min / max_diam)
        DI = min(DI, the_min)
    
    return DI
```
